```python
import pandas as pd
import numpy as np

df_freq = pd.read_csv('claims-cargo_Freq.csv', delimiter=';')
df_sev = pd.read_csv('claims-cargo_SEV.csv', delimiter=';')
```

```python
#frequency dataset

df_freq_sorted = df_freq.sort_values(by='cargo_type')

df_freq_sorted1 = df_freq_sorted.copy()
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^cobalt.*', 'cobalt', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^gold.*', 'gold', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^lithium.*', 'lithium', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^platinum.*', 'platinum', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^rare earth.*', 'rare earth', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^supplies.*', 'supplies', regex=True)
df_freq_sorted1['cargo_type'] = df_freq_sorted1['cargo_type'].str.replace('(?i)^titanium.*', 'titanium', regex=True)

df_freq_sorted = df_freq_sorted1

print(df_freq_sorted['cargo_type'].unique())



 
```

```python
# print rows where cargo_type is missing
df_freq_NA_cargo = df_freq_sorted[df_freq_sorted['cargo_type'].isna()]

df_freq_sorted[df_freq_sorted['cargo_type'].isna()].to_csv('missing_cargo_types.csv', index=False)
```
